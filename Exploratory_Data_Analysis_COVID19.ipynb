#***Top 5 Countries with the most COVID cases***

```python
# Deleting columns to clean the data.

df = df.drop([len(df)-1])
country_df = df.drop([0])
```

```python
# Defining the time space of the data.

today = datetime.now()
today_str = "%s %d, %d" % (date.today().strftime("%b"), today.day, today.year)
yesterday_str = "%s %d, %d" % (date.today().strftime("%b"), today.day-1, today.year)
clean = True
```

***Countries with the most Distribution of Total Cases***

```python
# Displaying the Pie chart to display the distribution of total cases.

fig = px.pie(country_df.iloc[0:20],
       names='Country',
       values='Total Cases',
      template='plotly_dark',
      title='Distribution of Total Cases ')
fig.show(renderer="colab")
```

```python
# Column chart of the countries with the most covid cases (cases recovered, new cases, deaths etc.)

country_l = country_df.columns[1:13]
LOOK_AT = 5 
fig = go.Figure()
c = 0
for i in country_df.index:
    if c < LOOK_AT:
        fig.add_trace(go.Bar(name=country_df['Country'][i], x=country_l, y=country_df.loc[i][1:13]))
    else:
        break
    c += 1
    
fig.update_layout(title={'text': f'{LOOK_AT} Countries with Most COVID Cases on %s' % yesterday_str, 'x': 0.5,
                         'xanchor': 'center', 'font': {'size': 20}}, yaxis_title="Percentage", yaxis_type="log", xaxis_tickangle=-90)
fig.show(renderer="colab")
```
